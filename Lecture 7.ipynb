```python
def greet(*names):
    for name in names:
        print("Hello",name)
greet('Anand','Arpit','Dayanand','Priya','Sapna', 'Rajneesh')
```

```python
def add(*x):
    s=0
    for _ in x:
        s+=_
    return s
add(450,56543,24374,787,456,4577)
```

```python
def fact(n):
    if n<0:
        return None
    elif n==0:
        return 1
    else:
        return n*fact(n-1)
print(fact(0))                                    #recursion
print(fact(-7))
print(fact(5))
print(fact(10))
```
