```python
def find_pairs_of_numbers(num_list, n):
    a=len(num_list)
    count=0
    for i in range(a):
        for j in range(i+1, a):
            if num_list[i]+num_list[j]==n:
                count+=1
    return count
find_pairs_of_numbers([1, 5, 8, 3, 4, 7, 9, 2], 6)
```

```python
def check_palindrome(str1):
    if str1==str1[::-1]:
        return True
    return False
```

```python
check_palindrome('abcba')
```

```python
check_palindrome('12121')
```

```python
check_palindrome('321234')
```

```python
check_palindrome('malayalam')
```

```python
def check_common(str1, str2):
    out_=''
    for i in str1:
        if i in str2:
            out_+=i
    return out_
```

```python
check_common('Anand','Priya')
```

```python
check_common('I want to learn Python.','Python is a funny language.')
```
