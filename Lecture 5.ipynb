```python
s=input()
out_=''
for i in s:
    if i.islower():
        out_+=i.upper()
    elif i.isupper():
        out_+=i.lower()
    else:
        out_+=i
    #print(out_)
print(out_)
```

```python
str1=input()
lst=str1.split(",")
num1=int(lst[0])
num2=int(lst[1])
num3=int(lst[2])
if num1+num2>num3 and num2+num3>num1 and num1+num3>num2:
    print('success')
else:
    print('failure')
```
